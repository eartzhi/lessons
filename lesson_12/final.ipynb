```python
import pandas as pd
import re 
```

```python
ratings_movies = pd.read_csv('data/ratings_movies.csv',index_col=0)
ratings_movies.head()
```

```python
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
ratings_movies.head()
```

```python
ratings_movies['year_release'] = ratings_movies['title'].apply(func=get_year_release)
```

Найдите сочетание жанров (genres) за 2018 год, которое имеет наибольший средний рейтинг (среднее по столбцу rating), и при этом число выставленных ему оценок (количество значений в столбце rating) больше 10.
Запишите сочетание так же, как оно указано в таблице (через разделитель |, без пробелов).

```python
filtered_mask = ratings_movies['year_release'] == 2018
genre_count = ratings_movies[filtered_mask].groupby(['genres'])['rating'].agg(['mean', 'count'])
filtered_mask_2 = genre_count['count'] > 10
genre_count[filtered_mask_2].nlargest(1, columns='mean')
```
