```python
import pandas as pd
```

```python
with open('data/dst-3.0_16_1_hh_database.csv', 'r', encoding='UTF-8') as input_f:
    with open('data/hh_database.csv', 'w', encoding='UTF-8') as output_f:
        for line in input_f:
            line = line.replace(' ,  ', ';')
            line = line.replace(' , ', ';')
            # line = line.replace(', ', ';')
            output_f.write(line)
```

```python
data = pd.read_csv('data/dst-3.0_16_1_hh_database.csv', sep=';')
data.head(2)
```

```python
data['Пол']=data['Пол, возраст'].apply(lambda x: x.split()[0])
data['Возраст']=data['Пол, возраст'].apply(lambda x: x.split(', ')[1])
data['Дата рождения']=data['Пол, возраст'].apply(lambda x: x.split(', ')[2])
data = data.drop('Пол, возраст', axis=1)

data['Город']=data['Город, переезд, командировки'].apply(lambda x: x.split(', ')[0])
data['Переезд']=data['Город, переезд, командировки'].apply(lambda x: x.split(', ')[1])
data['Командировки']=data['Город, переезд, командировки'].apply(lambda x: x.split(', ')[2])
data = data.drop('Город, переезд, командировки', axis=1)

data.head(2)
```

```python
data.describe()
```

```python
# data['Опыт работы']=data['Опыт работы'].apply(lambda x: x.replace('Опыт работы ', ''))
data['ЗП']=data['ЗП'].apply(lambda x: x.split(' ')[0])
```

```python
data['ЗП']=data['ЗП'].astype(int)
```

```python
data['Обновление резюме']=pd.to_datetime(data['Обновление резюме'], dayfirst=True)
```
